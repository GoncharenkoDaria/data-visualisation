Создайте синтетический набор данных с коррелированными объектами. Визуализируйте корреляционную матрицу набора данных с помощью тепловой карты.

```python
np.random.seed(42)

feature_1 = np.random.normal(50, 10, 100)  
feature_2 = feature_1 * 0.7 + np.random.normal(5, 2, 100)  
feature_3 = feature_1 * -0.5 + np.random.normal(10, 5, 100) 
feature_4 = np.random.normal(40, 15, 100)  

df = pd.DataFrame({
    'Feature_1': feature_1,
    'Feature_2': feature_2,
    'Feature_3': feature_3,
    'Feature_4': feature_4
})

correlation_matrix = df.corr()
df.head(2)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title("Тепловая карта корреляции")
plt.show()
```

Используя данный набор данных, сгенерировать трехмерную точечную диаграмму для визуализации распределения точек данных в трехмерном пространстве.


```python
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)
df.head(2)
```

```python
fig = px.scatter_3d(
    df,
    x='X', 
    y='Y', 
    z='Z',
    color='Z', 
    title='3D Scatter Plot of Data Points',
    labels={'X': 'X-axis', 'Y': 'Y-axis', 'Z': 'Z-axis'}
)

# Настраиваем оси
fig.update_layout(scene=dict(
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    zaxis_title='Z-axis',
    aspectmode='auto'  # Автоматический масштаб осей
))

# Показываем график
fig.show()
```

Используя данные о продажах, создайте тепловую карту, чтобы наглядно представить изменения в продажах в разные месяцы и дни.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(data)
df_sales.head(5)
```

```python

heatmap_data = df_sales.pivot_table(
    index='Day', 
    columns='Month', 
    values='Sales', 
    aggfunc='mean'
)

fig = go.Figure(
    data=go.Heatmap(
        z=heatmap_data.values,
        x=heatmap_data.columns,
        y=heatmap_data.index,
        colorscale='Viridis',
        colorbar_title='Avg Sales'
    )
)

fig.update_layout(
    title='Изменение объема продаж по Месяцам и дням',
    xaxis_title='Месяц',
    yaxis_title='День'
)

fig.show()
```

Используя заданные данные x и y, создайте трехмерный график поверхности для визуализации функции Z=sin( X 2+Y2)

```python

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
df.head(2)
```

```python
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.update_layout(
    title='3D Surface Plot: Z = sin(√(X² + Y²))',
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    )
)

fig.show()
```
