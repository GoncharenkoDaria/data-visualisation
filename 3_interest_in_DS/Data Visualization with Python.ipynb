Был проведен опрос, чтобы оценить интерес аудитории к различным темам Data Science, а именно::

    Большие данные (Spark / Hadoop)
    Анализ данных / статистика
    Журналистика данных
    Визуализация данных
    Глубокое обучение
    Машинное обучение

У участников было три варианта ответа на каждую тему: Очень интересно, в некоторой степени интересно и неинтересно. Опрос завершили 2233 респондента.

Результаты опроса были сохранены в csv-файле и доступны по ссылке: Topic_Survey_Assignment.

```python

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python

survey_responses = pd.read_csv('Topic_Survey_Assignment.csv')
print('Датасэт успешно загружен.')
```

```python
survey_responses.head(10)
```

```python
survey_responses.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
survey_responses.head(6)
```


Процент интереса респондентов к различным темам, связанным с наукой о данных.

```python

survey_percentage = survey_responses.div(2233).mul(100).round(decimals=2)
survey_percentage.head(10)
```

```python

ax = survey_percentage.plot(kind='bar', 
                           figsize=(20,8),
                           width=.8,
                           color=['#5cb85c', '#5bc0de', '#d9534f'])

plt.legend(loc='upper right', fontsize=14)
plt.xticks(fontsize=14)

ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize=16)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# remove y-axis ticks
ax.get_yaxis().set_ticks([])

rects = ax.patches

# create the labels from the dataframe by transposing and flattening
labels = survey_percentage.transpose().values.flatten()

for rect, label in zip(rects, labels):
    width = rect.get_width()
    height = rect.get_height()
    x = rect.get_x()
    y = rect.get_y() + rect.get_height() + 30
    ax.text(x + width/2, height+2, label, ha='center', va='bottom', fontsize=14)
    
```
